```python
def daily_variance(df):
    '''
    purpose: will rountine will take in a pandas dataframe and plot a column of data a key
             column set up for a groupby statement.
    input:   - standard stock data paired down to date, close and ticker_symbol
             - more generally:
                 - a pandas dataframe with a data column
                 - a key column for groupby
    output:  a seaborn voilinplot with n columns being the number of groups
             a dataframe showing the min, max and mean of the column data
    '''
 
    import pandas as pd
    import seaborn as sns
    from matplotlib import pyplot as plt
    # plot daily stock Variances

    dims = (15, 8)
    fig, ax = plt.subplots(figsize=dims)
    plt.suptitle("Daily Variance in Stocks", fontsize=15)
    plt.ylabel("close", fontsize=18)
    ax = sns.violinplot(x='Name', y='close', data=df)
    plt.xlabel("Company", fontsize=18)
    plt.show();

    min_df=df.groupby('Name')['close'].min().reset_index().rename(columns={'close': 'min'})
    max_df=df.groupby('Name')['close'].max().reset_index().rename(columns={'close': 'max'})
    mean_df=df.groupby('Name')['close'].mean().reset_index().rename(columns={'close': 'mean'})
    display(min_df.merge(max_df,on='Name').merge(mean_df,on='Name'))
    return



```

```python
df=pd.read_csv('all_stocks_5yr.csv')
name_lst=list(df.Name.unique()[:5])
df=df[df['Name'].isin(name_lst)]
daily_variance(df)
```
